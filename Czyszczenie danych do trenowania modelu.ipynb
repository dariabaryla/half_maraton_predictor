```python
import pandas as pd
```

```python
df_2023 = pd.read_csv('halfmarathon_wroclaw_2023__final.csv', sep=';')
df_2023.head()
```

```python
df_2024 = pd.read_csv('halfmarathon_wroclaw_2024__final.csv', sep=';')
df_2024.head()
```

```python
df_2023.info()
```

```python
df_2024.info()
```

```python
# Czyszczenie danych
# Zostawiam tylko potrzebne kolumny: Płeć, Rocznik, 5 km Czas, Czas
needed_columns = ['Płeć', 'Rocznik', '5 km Czas', 'Czas']
df_2023_clean = df_2023[needed_columns].copy()
df_2024_clean = df_2024[needed_columns].copy()
```

```python
# Usuwanie wierszy z brakującymi danymi (NaN)
df_2023_clean = df_2023_clean.dropna()
df_2024_clean = df_2024_clean.dropna()
```

```python
# Usuwanie wierszy z DNS, DNF lub innymi niecyfrowymi wartościami w czasach
df_2023_clean = df_2023_clean[~df_2023_clean['Czas'].isin(['DNS', 'DNF', ''])]
df_2024_clean = df_2024_clean[~df_2024_clean['Czas'].isin(['DNS', 'DNF', ''])]
df_2023_clean = df_2023_clean[~df_2023_clean['5 km Czas'].isin(['DNS', 'DNF', ''])]
df_2024_clean = df_2024_clean[~df_2024_clean['5 km Czas'].isin(['DNS', 'DNF', ''])]
```

```python
# Obliczenie wieku na podstawie roku wydarzenia i Rocznik
df_2023_clean['Wiek'] = 2023 - df_2023_clean['Rocznik'].astype(float).astype(int)
df_2024_clean['Wiek'] = 2024 - df_2024_clean['Rocznik'].astype(float).astype(int)
```

```python
# Konwersja czasów na sekundy (z formatu HH:MM:SS na całkowite sekundy)
def time_to_seconds(time_str):
    if pd.isna(time_str) or time_str in ['DNS', 'DNF']:
        return None
    try:
        h, m, s = map(int, time_str.split(':'))
        return h * 3600 + m * 60 + s
    except:
        return None

df_2023_clean['5km_seconds'] = df_2023_clean['5 km Czas'].apply(time_to_seconds)
df_2023_clean['total_seconds'] = df_2023_clean['Czas'].apply(time_to_seconds)
df_2024_clean['5km_seconds'] = df_2024_clean['5 km Czas'].apply(time_to_seconds)
df_2024_clean['total_seconds'] = df_2024_clean['Czas'].apply(time_to_seconds)
```

```python
# Usuwanie wierszy, gdzie konwersja czasów nie powiodła się (None)
df_2023_clean = df_2023_clean.dropna(subset=['5km_seconds', 'total_seconds', 'Wiek'])
df_2024_clean = df_2024_clean.dropna(subset=['5km_seconds', 'total_seconds', 'Wiek'])
```

```python
# Kodowanie płci: M -> 0, K -> 1 (dla modelu)
df_2023_clean['Płeć_encoded'] = df_2023_clean['Płeć'].map({'M': 0, 'K': 1})
df_2024_clean['Płeć_encoded'] = df_2024_clean['Płeć'].map({'M': 0, 'K': 1})
```

```python
# Usuwanie oryginalnych kolumn czasów i Płeć
df_2023_clean = df_2023_clean[['Płeć_encoded', 'Wiek', '5km_seconds', 'total_seconds']]
df_2024_clean = df_2024_clean[['Płeć_encoded', 'Wiek', '5km_seconds', 'total_seconds']]
```

```python
# Wyświetlenie wyczyszczonych danych
print("Wyczyszczone dane 2023:")
display(df_2023_clean.head())
print("Wyczyszczone dane 2024:")
display(df_2024_clean.head())
```

```python
# Zapisanie wyczyszczonych danych do nowych plików CSV
df_2023_clean.to_csv('clean_halfmarathon_2023.csv', index=False)
df_2024_clean.to_csv('clean_halfmarathon_2024.csv', index=False)
```
